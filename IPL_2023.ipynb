```python
import pandas as pd
```

# Results

```python
df_results=pd.read_csv('Results.csv')
df_results.head()
```

```python
df_results['match']=df_results['Team 1'] + ' vs ' + df_results['Team 2']
df_results.head()
```

```python
df_results['match_id']=''
for index, row in df_results.iterrows():
    df_results.at[index, 'match_id'] = index + 1
    
df_results.head()
```

```python
df_new_results=df_results
df_new_results.drop(columns=['Scorecard', 'Team 1', 'Team 2'], inplace=True)
df_new_results.head()
```

```python
df_new_results.shape
```

```python
df_new_results.to_csv('results_updated.csv')
```

#  For Batting

```python
df_batting= pd.read_csv('batting_summary.csv',encoding='windows-1254')
df_batting.head()
```

```python
df_batting['SR']=(df_batting['Run']/df_batting['Ball'])*100
df_batting.head()
```

```python
df_batting['SR']=df_batting['SR'].round(2)
df_batting.head(10)
```

```python
new_batting=df_batting
new_batting.head()
```

```python
new_batting['out/not_out']=new_batting.out_by.apply(lambda x: 'not_out' if 'Not Out' in x else 'out')
new_batting.head(11)
```

```python
new_batting.drop(columns='out_by',inplace=True)
new_batting.head()
```

```python
new_batting.isnull().sum()
```

```python
new_batting.rename(columns={'match_no': 'match_id'},inplace=True)
new_batting.head()
```

```python
new_batting.head(20)
new_batting.to_csv('batting_updated.csv')
```
